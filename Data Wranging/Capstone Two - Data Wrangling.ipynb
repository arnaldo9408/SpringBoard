# Convert all the date's columns to datetime for future use

```python
df_features['Date'] =  pd.to_datetime(df_features['Date'])
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])
df_stock_price['Date'] = pd.to_datetime(df_stock_price['Date'])
```

# Start the merging of all the data frames to have all the information in one big data frame

#### By looking the data, we are taking off two data frames since there is no concordance with the dates of the other data frames. Columns we are disregarding are test and store

```python
'''We start merging the features data frame with the train dataframe. Left Join to include values, and on the date 
column as the concordance column between the two data frames'''
df_walmart = pd.merge(df_features, df_train, on = 'Date', how = 'left')
```

```python
'''We continued merging the df_walmart data frame with the test dataframe. Left Join to include values and the
date column  as the concordance column between the two data frames'''
df_walmart = pd.merge(df_walmart, df_test, on = 'Date', how = 'left')
```

```python
'''We continued merging the df_walmart data frame with the stock price dataframe. Left Join to include values and the
date column  as the concordance column between the two data frames'''
df_walmart = pd.merge(df_walmart,df_stock_price,on = 'Date', how = 'left')
```

```python
#We want to remove some duplicates. Identified the IsHoliday colum as a duplicate column
df_walmart = df_walmart.drop(columns = ['IsHoliday_y', 'IsHoliday'])
```

```python
#See the first 5 rows of the new data frame
df_walmart.head()
```

```python
#See the data frame types to confirm all is in order
df_walmart.info()
```

```python
#Check Null values in order to fill them out
df_walmart.isna().sum()
```

# Fill of the NaN Values

```python
'''Replace the NaN Values
Close price: It should have been a weekend day where the stock is not traded. Replace NaN values with the value
of the days before
Unemployment: Replace this value with the mean
Weekly Sales
CPI: Fill NaN values with the mean
Weekly_Sales = Fill NaN values with the mean
Store_y, Dept_y, Store, and Dept_x can be deleted from the data Frame
Markdown : Replace NaN with cero (no discounts)
'''
```

```python
#Extract columns I am interested of the data frame
df_walmart_sales = df_walmart[['Date','Temperature','Fuel_Price','MarkDown1', 'MarkDown2', 'MarkDown3', 
                               'MarkDown4', 'MarkDown5', 'CPI', 'Unemployment', 'IsHoliday_x', 'Weekly_Sales',
                              'Close']]
```

```python
#Replace Markdown NaN values
df_walmart_sales['MarkDown1'].fillna(0, inplace = True)
df_walmart_sales['MarkDown2'].fillna(0, inplace = True)
df_walmart_sales['MarkDown3'].fillna(0, inplace = True)
df_walmart_sales['MarkDown4'].fillna(0, inplace = True)
df_walmart_sales['MarkDown5'].fillna(0, inplace = True)
```

```python
#Replace Close price using backfill methid
df_walmart_sales['Close'].fillna(method = 'bfill', inplace = True)
```

```python
#Replace Unemployment with the mean
df_walmart_sales['Unemployment'].fillna(df_walmart_sales['Unemployment'].mean(), inplace = True)
```

```python
# Fill CPA values with the mean
df_walmart_sales['CPI'].fillna(df_walmart_sales['CPI'].mean(), inplace = True)
```

```python
#Fill Weekly Sales with the mean
df_walmart_sales['Weekly_Sales'].fillna(df_walmart_sales['Weekly_Sales'].mean(), inplace = True)
```

```python
#Check the fill of the NaN values went through
df_walmart_sales.isna().sum()
```

```python
#Rename some columns
df_walmart_sales.rename(columns = {'IsHoliday_x':'IsHoliday'}, inplace = True) 
```

# Grouped of the data frame to have it more readable and easier to work on

```python
'''We saw the date columm for some reason during the merge, it has repetition on the values, therefore we are able to
group the data in dates. In addition, since all values are repetions, we are also able to aggregate them as mean
and final results would be summarized and not affected'''
```

```python
df_walmart_sales_grouped = df_walmart_sales.groupby('Date').agg('mean')
```
