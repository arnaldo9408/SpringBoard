# Convert all the date's columns to datetime for future use

```python
df_features['Date'] =  pd.to_datetime(df_features['Date'])
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])
df_stock_price['Date'] = pd.to_datetime(df_stock_price['Date'])
```

# Start the merging of all the data frames to have all the information in one big data frame

#### By looking the data, we are taking off two data frames since there is no concordance with the dates of the other data frames. Columns we are disregarding are test and store

```python
'''We start merging the features data frame with the train dataframe. Left Join to include values, and on the date, store and Is Holiday 
column as the concordance column between the two data frames'''
df_features_train = pd.merge(df_train, df_features,how = 'inner', on = ['Date', 'Store', 'IsHoliday'])
```

```python
df_features_train.head()
```

```python
df_features_train.head()
```

```python
'''We continued merging the df_walmart data frame with the stock price dataframe. Left Join to include values and the
date column  as the concordance column between the two data frames'''
df_features_train_stock = pd.merge(df_features_train, df_stock_price,how = 'inner', on = ['Date'])
```

```python
#See the new dataframe 
df_features_train_stock.head()
```

```python
#See the data frame types to confirm all is in order
df_features_train_stock.info()
```

# Fill of the NaN Values

```python
'''Replace the NaN Values
Close price: It should have been a weekend day where the stock is not traded. Replace NaN values with the value
of the days before
Unemployment: Replace this value with the mean
Weekly Sales
CPI: Fill NaN values with the mean
Weekly_Sales = Fill NaN values with the mean
Store_y, Dept_y, Store, and Dept_x can be deleted from the data Frame
Markdown : Delete columns since the percentage of NaN values is too high
'''
```

```python
df_features_train_stock.columns
```

```python
#Extract columns I am interested of the data frame
df_features_train_stock = df_features_train_stock[['Date','Temperature','Fuel_Price', 'CPI', 'Unemployment', 'IsHoliday', 'Weekly_Sales',
                              'Close', 'Dept']]
```

```python
df_features_train_stock.head()
```

```python
df_features_train_stock.tail()
```

```python
#Replace Close price using backfill methid
df_features_train_stock['Close'].fillna(method = 'bfill', inplace = True)
```

```python
#Replace Unemployment with the mean
df_features_train_stock['Unemployment'].fillna(df_features_train_stock['Unemployment'].mean(), inplace = True)
```

```python
# Fill CPA values with the mean
df_features_train_stock['CPI'].fillna(df_features_train_stock['CPI'].mean(), inplace = True)
```

```python
df_features_train_stock.dropna()
```

## Change the department to a categorical variable

```python
#Aggregate the department as a dummy variable to get it as a possible exploratory variable
department_dummies = pd.get_dummies(df_features_train_stock['Dept'])
```

```python
#We no longer need the department column since we are going to replace it with the categorical variable
df_features_train_stock.drop(columns = 'Dept', inplace = True)
```

```python
#Concat the categorical data frame with the df 
df = pd.concat([df_features_train_stock, department_dummies], axis = 1)
```
